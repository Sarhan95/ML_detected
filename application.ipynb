```python
res=pd.DataFrame(columns=["var","statistic_test","p_value"])

for var in varss:
    class_fraud=data.loc[data["Class"]==1,var]
    class_nonfraud=data.loc[data["Class"]==0,var]
    stat,p_values=mannwhitneyu(class_fraud,class_nonfraud,alternative="two-sided")

    res.loc[len(res)]=[var,stat,p_values]
    
    
   
res

```

```python
######## apprentissage automatique d√©ploiment des mod√©les
######## divisition notre dataset
x=data.drop("Class",axis=1)
y=data["Class"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=123)
print("dim de x_train:",x_train.shape)
print("dim de x_test:",x_test.shape)
print("dim de y_test:",y_test.shape)
print("dim de y_train:",y_train.shape)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,f1_score
from sklearn.model_selection import learning_curve


########  regression logistique
rl=LogisticRegression(random_state=123,max_iter=1000)
###### creation une fonction #########
def evaluation(rl):
    rl.fit(x_train,y_train)
    y_pred=rl.predict(x_test)
    print(confusion_matrix(y_test,y_pred))
    print(classification_report(y_test,y_pred))
    N,train_scores,val_scores=learning_curve(rl,x_train,y_train,cv=5,scoring="f1",train_sizes=np.linspace(0.1,1,10))
    plt.figure(figsize=(10,20))
    plt.plot(N,train_scores.mean(axis=1),label="train_score")
    plt.plot(N,val_scores.mean(axis=1),label="val_score")
    plt.show()
    
    
    
```

```python
evaluation(rl)
```
